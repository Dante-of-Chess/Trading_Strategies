```python
import pandas as pd
from pandas.tseries.holiday import USFederalHolidayCalendar
from datetime import datetime, timedelta

def get_us_working_days(start_year, end_year):
    all_working_days = []

    for year in range(start_year, end_year + 1):
        # Create a US Federal Holiday calendar
        cal = USFederalHolidayCalendar()

        # Get all holidays for the specified year
        holidays = cal.holidays(datetime(year, 1, 1), datetime(year, 12, 31))

        # Generate a range of dates for the specified year
        start_date = datetime(year, 1, 1)
        end_date = datetime(year, 12, 31)
        date_range = pd.date_range(start_date, end_date)

        # Filter out weekends (Saturday and Sunday) and holidays
        working_days = [date for date in date_range if date.weekday() < 5 and date not in holidays]

        all_working_days.extend(working_days)

    return all_working_days

# Example: Get all US working days between 2005 and 2023
start_year = 2005
end_year = 2024
us_working_days_2005_to_2023 = get_us_working_days(start_year, end_year)
print(us_working_days_2005_to_2023)
dates_to_remove = ['2015-04-03', '2016-03-25', '2017-04-14', '2018-03-30', '2021-04-02', '2022-04-15', '2023-04-07']

# Using a list comprehension
us_working_days_2005_to_2023 = [date for date in us_working_days_2005_to_2023 if date not in dates_to_remove]

# Alternatively, using filter and lambda
us_working_days_2005_to_2023 = list(filter(lambda date: date not in dates_to_remove, us_working_days_2005_to_2023))

```
